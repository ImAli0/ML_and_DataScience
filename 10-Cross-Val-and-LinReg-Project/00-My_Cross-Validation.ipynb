```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Advertising.csv')
```

```python
df.head()
```

```python
# CREATE X and y
X = df.drop('sales', axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

# SCALE THE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=3)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```

```python
model = Ridge(alpha=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mean_squared_error(y_test, y_pred)
```

## Train | Validation | Test

```python
# SPLIT TWICE

X_train, X_other, y_train, y_other = train_test_split(X, y, test_size=0.33, random_state=0)
X_eval, X_test, y_eval, y_test = train_test_split(X_other, y_other, test_size=0.33, random_state=0)
```

```python
# SCALE THE DATA
scaler=StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_eval = scaler.transform(X_eval)
X_test = scaler.transform(X_test)
```

```python
model = Ridge(alpha=2)
model.fit(X_train, y_train)
y_pred = model.predict(X_eval)

mean_squared_error(y_eval, y_pred)
```

```python
model = Ridge(alpha=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_eval)

mean_squared_error(y_eval, y_pred)
```

```python
y_final_test_pred = model.predict(X_test)

mean_squared_error(y_test, y_final_test_pred)
```

# Cross Validation with cross_val_score

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
model = Ridge(alpha=2)

# SCORING
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
scores
```

```python
abs(scores.mean())
```

```python
model = Ridge(alpha=1)

scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
abs(scores.mean())
```

# Final Evaluation

```python
model.fit(X_train, y_train)
y_final_pred = model.predict(X_test)

mean_squared_error(y_test, y_final_pred)
```

# Cross Validation with cross_validate

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
model = Ridge(alpha=2)

from sklearn.model_selection import cross_validate
scores = cross_validate(model, X_train, y_train, scoring=['neg_mean_absolute_error', 'neg_mean_squared_error', 'max_error'], cv=10)
scores = pd.DataFrame(scores)
scores
```

```python
scores.mean()
```

```python
model = Ridge(alpha=1)
scores = cross_validate(model, X_train, y_train, scoring=['neg_mean_absolute_error', 'neg_mean_squared_error', 'max_error'], cv=10)
```

```python
pd.DataFrame(scores).mean()
```

# Final Evaluation

```python
model.fit(X_train, y_train)
y_final_pred = model.predict(X_test)

mean_squared_error(y_test, y_final_pred)
```
