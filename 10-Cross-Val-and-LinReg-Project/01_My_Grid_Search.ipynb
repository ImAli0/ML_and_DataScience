```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

# Data

```python
df = pd.read_csv('/content/Advertising.csv')
df.head()
```

# Formatting Data

```python
# CREATE X AND y
X = df.drop('sales', axis=1)
y = df['sales']

# TRAIN TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

#  SCALE DATA
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Model

```python
base_elastic_model = ElasticNet()
```

# Grid Search

```python
param_grid = {'alpha':[0.1,0.2,0.3,0.5,0.7,1,2,3,5,7,10], 'l1_ratio':[.1,.2,.3,.5,.7,.8,.9,.95,.99,1]}
grid_model = GridSearchCV(estimator=base_elastic_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=10, verbose=3)
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
cv_results = pd.DataFrame(grid_model.cv_results_)
cv_results
```

# Using Best Model From Grid Search

```python
y_pred = grid_model.predict(X_test)
mean_squared_error(y_test, y_pred)
```
