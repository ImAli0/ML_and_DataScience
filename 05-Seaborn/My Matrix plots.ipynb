```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('country_table.csv')
```

```python
df
```

```python
df=df.set_index('Countries')
```

```python
df
```

```python
sns.heatmap(df)
```

```python
rates=df.drop('Life expectancy',axis=1)
```

```python
sns.heatmap(rates)
```

```python
sns.heatmap(rates,linewidth=0.4)
```

```python
sns.heatmap(rates,linewidth=0.6,annot=True)
```

```python
sns.heatmap(rates,linewidth=0.4,annot=True,cmap='viridis')
```

```python
sns.heatmap(rates,linewidth=0.4,annot=True,cmap='crest_r')
```

```python
sns.clustermap(rates)
```

```python
sns.clustermap(rates,col_cluster=False)
```

```python
sns.clustermap(rates,col_cluster=False,figsize=(12,8),cbar_pos=(-0.1,.2,.04,
                                                              .5))
```
