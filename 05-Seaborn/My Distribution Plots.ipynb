### Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df.info()
```

### Rugplot

```python
# The y axis doesn't really represent anything
# X axis is just a stick per data point
sns.rugplot(x='salary',data=df)
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

### displot() and histplot()

```python
sns.displot(data=df,x='salary',kde=True)
```

### Focusing on the Histogram

```python
sns.displot(x='salary',data=df)
```

```python
sns.histplot(x='salary',data=df)
```

### Number of Bins

```python
sns.histplot(data=df,x='salary',bins=15)
```

```python
plt.figure(figsize=(12,8))
sns.histplot(data=df,x='salary',bins=100)
```

## Adding in a grid and styles

```python
plt.figure(figsize=(12,8))
sns.set(style='darkgrid')
sns.histplot(data=df,x='salary',bins=100,kde=True)
```

## Adding in keywords from matplotlib

```python
sns.displot(data=df,x='salary',bins=20,kde=True,color='darkgreen',
           edgecolor='darkred',lw=4,ls='--')
```

## The Kernel Density Estimation Plot

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
type(sample_ages)
```

```python
sample_ages=pd.DataFrame(sample_ages,columns=['age'])
```

```python
type(sample_ages)
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=15,rug=True)
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=15,rug=True,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='age')
```

## Cut Off KDE

```python
plt.figure(figsize=(12,8))
sns.kdeplot(data=sample_ages,x='age',clip=[0,100])
```

### Bandwidth

```python
sns.kdeplot(data=sample_ages,x='age',bw_adjust=0.1)
```

```python
sns.kdeplot(data=sample_ages,x='age',bw_adjust=0.4)
```

```python
sns.kdeplot(data=sample_ages,x='age',bw_adjust=1)
```

### Basic Styling

```python
sns.kdeplot(data=sample_ages,x='age',bw_adjust=0.3,shade=True,color='green')
```
