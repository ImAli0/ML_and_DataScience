```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('application_record.csv')
```

```python
df
```

```python
employed=df
employed['DAYS_BIRTH']=-employed['DAYS_BIRTH']
employed['DAYS_EMPLOYED']=-employed['DAYS_EMPLOYED']
```

```python
employed
```

```python
employed=employed[employed['DAYS_EMPLOYED']>0]
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='DAYS_BIRTH',y='DAYS_EMPLOYED',data=employed,alpha=0.01,lw
               =0)
```

```python
df['Age in Years']=df['DAYS_BIRTH']/365
```

```python
df
```

```python
plt.figure(figsize=(8,4))
sns.histplot(x='Age in Years',data=df,color='r',lw=2,ec='black',
            alpha=0.9, bins=45)
```

```python
half=df.sort_values('AMT_INCOME_TOTAL',ascending=False)[219278:]
```

```python
half
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',data=half,hue='FLAG_OWN_REALTY')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,title='FLAG_OWN_REALTY')
plt.title('Income Totals per Family Status for Bottom Half of Earners')
```

```python
plt.figure(figsize=(12,5))

bottom_half_income = df.nsmallest(n=int(0.5*len(df)),columns='AMT_INCOME_TOTAL')
sns.boxplot(x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',data=bottom_half_income,hue='FLAG_OWN_REALTY')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,title='FLAG_OWN_REALTY')
plt.title('Income Totals per Family Status for Bottom Half of Earners')
```
