```python
g=sns.PairGrid(df)
g=g.map_upper(sns.scatterplot,lw=3)
g=g.map_diag(sns.kdeplot)
g=g.map_lower(sns.kdeplot,colors='black')
```

```python
g = sns.PairGrid(df, hue="gender", palette="viridis",hue_kws={"marker": ["o", "+"]})
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="r", s=40)
g = g.map_diag(sns.kdeplot,shade=True)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();

# Safely ignore the warning, its telling you it didn't use the marker for kde plot
```
