```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df.tail()
```

## Countplot()

```python
plt.figure(figsize=(10,5),dpi=200)
sns.countplot(x='division',data=df)
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.countplot(x='level of education',data=df)
```

## Breakdown within another category with 'hue'

```python
plt.figure(figsize=(10,5),dpi=200)
sns.countplot(x='level of education',data=df,hue='division')
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.countplot(x='level of education',data=df,hue='division',
              palette='Dark2')
```

# barplot()

```python
df['salary'].describe()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='division',y='salary',data=df,estimator=np.mean,ci='sd')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='division',y='salary',data=df,estimator=np.mean,ci='sd',
            hue='level of education')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='division',y='salary',data=df,estimator=np.mean,ci='sd',
            hue='level of education')
plt.legend(bbox_to_anchor=(1.03,1))
```
