```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df,hue=
           'gender')
plt.legend(bbox_to_anchor=(1.001,1))
```

```python
sns.boxplot(x='math score',y='parental level of education',data=df,orient='h')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df,hue=
           'gender',width=0.4)
plt.legend(bbox_to_anchor=(1.001,1))
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(y='parental level of education',x='math score',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,hue=
              'gender')
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,hue=
              'gender',split=True)

plt.legend(bbox_to_anchor=(1.1,1))
```
