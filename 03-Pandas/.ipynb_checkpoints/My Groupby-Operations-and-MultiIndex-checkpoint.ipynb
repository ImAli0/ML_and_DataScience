# Groupby Operations and Multi-level Index


```python
import numpy as np
import pandas as pd
```

## Data

```python
df=pd.read_csv('mpg.csv')
```

```python
df
```

## groupby() method

```python
df.groupby('cylinders')
```

```python
df.groupby('cylinders').mean()
```

```python
df.groupby('cylinders').mean().index
```

```python
df.groupby('cylinders').mean().columns
```

```python
df.groupby('cylinders').mean()['displacement']
```

```python
df.groupby('cylinders').mean().describe()
```

```python
df.groupby('cylinders').describe()
```

```python
df.groupby('cylinders').describe().transpose()
```

## Groupby Multiple Columns

```python
df
```

```python
df.groupby(['cylinders','model_year']).mean()
```

```python
df.groupby(['cylinders','model_year']).mean().index
```

# MultiIndex
## The MultiIndex Object

```python
df.groupby(['cylinders','model_year']).mean().index.levels
```

```python
df.groupby(['cylinders','model_year']).mean().index.names
```

# Indexing with the Hierarchial Index

```python
df.groupby(['cylinders','model_year']).mean().head()
```

## Grab Based on Outside Index

```python
df.groupby(['cylinders','model_year']).mean().loc[8]
```

```python
df.groupby(['cylinders','model_year']).mean().loc[[6,8]]
```

```python
df.groupby(['cylinders','model_year']).mean().loc[(8,78)]
```

# Grab Based on Cross-section with .xs() 

```python
df
```

```python
df=df.groupby(['cylinders','model_year']).mean()
```

```python
df
```

```python
df.groupby(['cylinders','model_year']).mean().loc[8]
```

```python
df.xs(key=8,axis=0,level='cylinders')
```

```python
df[df['model_year'].isin([70,72,74,76,78,80])].groupby(['cylinders','model_year']).mean()
```

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

```python
df
```

## Swap levels

```python
df=df.swaplevel()
```

```python
df
```

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

```python
df
```

## Sorting MultiIndex

```python
df.sort_index(level='cylinders',ascending=False)
```

```python
df.sort_index(level='model_year',ascending=True)
```

# Advanced: agg() method

```python
df=pd.read_csv('mpg.csv')
```

```python
df
```

```python
df.agg(['max','mean'])
```

### Specify aggregate methods per column

```python
df.agg({'mpg':['std','mean'],'displacement':['min','max','mean']})
```

```python
df
```

```python
df.groupby('cylinders').agg({'mpg':['std','mean'],'displacement':['min','max','mean']})
```
