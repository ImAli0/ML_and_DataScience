# Principal Component Analysis

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('/content/cancer_tumor_data_features.csv')
df.head()
```

## Data Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
X_scaled
```

## SKLearn Implementation

```python
pca = PCA(n_components=3)
principal_components = pca.fit_transform(X_scaled)
```

```python
fig = plt.figure(figsize=(20,12))
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter points
ax.scatter(principal_components[:, 0], principal_components[:, 1], principal_components[:, 2], c='blue', marker='o')

# Add labels and title (optional)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D Scatter Plot of 3 PCs')

plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dictionary = load_breast_cancer()
cancer_dictionary.keys()
```

```python
cancer_dictionary['target']
```

```python
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter points
ax.scatter(principal_components[:, 0], principal_components[:, 1], principal_components[:, 2], c=cancer_dictionary['target'], marker='o')

# Add labels and title (optional)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D Scatter Plot of 3 PCs')

plt.show()
```

```python
pca.n_components
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, index=['PC1', 'PC2', 'PC3'], columns=df.columns)
df_comp
```

```python
plt.figure(figsize=(20,4), dpi=200)
sns.heatmap(df_comp, annot=True)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
explained_variance = []

for n in range(1,30):
  pca = PCA(n)
  pca.fit(X_scaled)

  explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30), explained_variance)
plt.xlabel('Number of components')
plt.ylabel('Explained_variance')
```
