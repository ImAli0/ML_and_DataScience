## Plotting on axes objects

Just as before, we simple .plot() on the axes objects, and we can also use the .set_ methods on each axes.

Let's explore this, make sure this is all in the same cell:

```python
fig, axes=plt.subplots(nrows=1,ncols=2)

for axe in axes:
    axe.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b);
```

```python
fig, axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

```python
fig, axes=plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)

plt.tight_layout()
```

```python
f, axes=plt.subplots(nrows=2,ncols=2,figsize=(12,8))
axes

axes[0][0].plot(b,a)
axes[0][0].set_title('0 0 Title')

axes[1][1].plot(x,y)
axes[1][1].set_title('1 1 Title');
axes[1][1].set_xlabel('1 1 X label')

axes[0][1].plot(y,x)
axes[1][0].plot(a,b)

f.suptitle("figure level")
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)


fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.1,)
```

## Exporting plt.subplots()

```python
# NOTE! This returns 2 dimensional array
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(15,8))

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.savefig('My subplots.png', bbox_inches='tight')
```
