### Basic Array Plot

Let's walk through a very simple example using two numpy arrays. You can also use lists, but most likely you'll be passing numpy arrays or pandas columns (which essentially also behave like arrays).

**The data we want to plot:**


```python
import numpy as np
```

```python
x=np.arange(0,10)
```

```python
y=2*x
```

# Using Matplotlib with plt.plot() function calls

## Basic Matplotlib Commands

We can create a very simple line plot using the following ( I encourage you to pause and use Shift+Tab along the way to check out the document strings for the functions we are using).

```python
plt.plot(x,y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
```

### Editing more figure parameters

```python
plt.plot(x,y)
plt.xlabel('X label Title Here')
plt.ylabel('Y label title Here')
plt.title('String Title Here')
plt.xlim(0,20) # Lower Limit, Upper Limit
plt.ylim(0,25) # Lower Limit, Upper Limit
```
