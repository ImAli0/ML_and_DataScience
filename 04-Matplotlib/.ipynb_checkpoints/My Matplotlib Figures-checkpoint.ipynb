### Matplotlib Object Oriented Method
Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.

## Creating a Figure

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

```python
# Create blank canvas
fig=plt.figure()
```

**NOTE: ALL THE COMMANDS NEED TO GO IN THE SAME CELL!**

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure (empty canvas)
fig=plt.figure()

# Add a set of axes to figure
axes=fig.add_axes([1,1,1,1]) # left, bottom, width, height (range 0 to 1)

#Plot on that set of axes
axes.plot(x,y)
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(a, b)

plt.show()
```

### Adding another set of axes to the Figure

So far we've only seen one set of axes on this figure object, but we can keep adding new axes on to it at any location and size we want. We can then plot on that new set of axes.

```python
type(fig)
```

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure. Note how we're plotting a,b twice here

```python
# Creates blank canvas
fig = plt.figure()

axes0=fig.add_axes([0,0,2,2]) # Large Figure
axes1=fig.add_axes([2.1,0.5,1,1]) # Smaller figure

# Larger figure axes 0
axes0.plot(a,b)

# Using set_ to add to the axes figure
axes0.set_xlabel('X label')
axes0.set_ylabel('Y label')
axes0.set_title('Large figure')

# Smaller figure axes 1
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('Smaller figure')
axes1.plot(a,b)
```

```python
# Creates blank canvas
fig=plt.figure()

axes0=fig.add_axes([0,0,2,2]) # Large figure
axes1=fig.add_axes([1.85,0.3,0.5,0.5]) # Small figure

# Large figure axes 0
axes0.plot(a,b)
axes0.set_xlabel('X label')
axes0.set_ylabel('Y label')
axes0.set_title('Large figure')

# Small figure
axes1.plot(a,b)
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('Zoomed in')
axes1.set_xlim(6,10)
axes1.set_ylim(5000,8000)

# Inserting  Axes 3 
axes2=fig.add_axes([1.87,0.95,0.5,0.5])
axes2.plot(a,b)
```

### Figure Parameters

```python
fig=plt.figure(figsize=(9,7),dpi=1000)

axes0=fig.add_axes([0,0,1,1])
axes0.plot(a,b)
```

## Exporting a Figure

```python
fig=plt.figure()

axes0=fig.add_axes([0,0,1,1])


axes0.plot(a,b)
axes0.set_xlabel('X')
fig.savefig('My_figure.png', bbox_inches='tight')
```

```python
fig=plt.figure(figsize=(10,7))

axes0=fig.add_axes([0,0,1,1]) # Full figure
axes1=fig.add_axes([-0.5,0.25,0.5,0.5]) # Small figure

axes0.plot(x,y)
axes1.plot(x,y)

fig.savefig('My test.png',bbox_inches='tight')
```
