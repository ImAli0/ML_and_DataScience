```python
import numpy as np
```

```python
m=np.linspace(0,10,11)

print(f"The array m should look like this: \n\n{m}")
```

```python
m
```

```python
c=299792458;
E=m*c**2
```

```python
print(f"The array E should look like this: \n\n{E}")
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(m,E,color='red',lw=5)
ax.set_xlabel('Mass in Grams')
ax.set_ylabel('Energy in Joules')
ax.set_title('E=mc^2')
plt.xlim(0,10);
```

```python
plt.yscale('log')
plt.plot(m,E,color='red',lw=5)
plt.xlim(0,10)
plt.grid(axis='y',which='both')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
plt.title('E=mc^2')
plt.show()
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(labels,july16_2007,label='july16_2007')
axes.plot(labels,july16_2020,label='july16_2020')
plt.legend();
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(labels,july16_2007,label='july16_2007')
axes.plot(labels,july16_2020,label='july16_2020')
plt.legend(loc=(1.04,0.5));

```

```python
fig, axes=plt.subplots(2,1,figsize=(12,8))

axes[0].plot(labels,july16_2007)
axes[1].plot(labels,july16_2020)
axes[0].set_title('July 16th, 2007')
axes[1].set_title('July 16th, 2020');
```

```python
fig, ax=plt.subplots(figsize=(12,8))

ax.plot(labels,july16_2007,label='july16_2007', color='blue', lw=2)
ax.set_ylabel('2007',fontsize=18,color='blue')
ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(4)

for label in ax.get_yticklabels():
    label.set_color('blue')
plt.yticks(fontsize=15)
ax0=ax.twinx()

ax0.plot(labels,july16_2020,label='july16_2020', c='r', lw=2)
ax0.set_ylabel('2020',fontsize=18,c='r')
ax0.spines['right'].set_color('red')
ax0.spines['right'].set_linewidth(4)

for label in ax0.get_yticklabels():
    label.set_color('red')
ax.set_title("July 16th Yield Curves");
plt.yticks(fontsize=15);
```
